```python
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt

my_image = Image.open('CR2.jpg')

sharp = my_image.filter(ImageFilter.SHARPEN)
sharp.save('D:/CR2S.jpg')
sharp.show()
plt.imshow(sharp)
plt.show()
```

```python
import matplotlib.pyplot as plt
img = Image.open('CR2.jpg')
plt.imshow(img)
plt.show()
flip = img.transpose(Image.FLIP_LEFT_RIGHT)

flip.save('D:/CR3Z.jpg')
plt.imshow(flip)
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
im = Image.open('CR2.jpg')

width, height = im.size

im1 = im.crop((280,100,800,600))

im1.show()
plt.imshow(im1)
plt.show()
```
